```python
import numpy
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
linear_bow_res = pandas.read_csv("data/result/train_result_2_linear_bow.csv")
linear_tfidf_res = pandas.read_csv("data/result/train_result_2_linear_tfidf.csv")
rbf_bow_res = pandas.read_csv("data/result/train_result_2_rbf_bow.csv")
rbf_tfidf_res = pandas.read_csv("data/result/train_result_2_rbf_tfidf.csv")
poly_bow_res = pandas.read_csv("data/result/train_result_2_poly_bow.csv")
poly_tfidf_res = pandas.read_csv("data/result/train_result_2_poly_tfidf.csv")
```

```python
plt.plot(
    linear_bow_res['C'],
    linear_bow_res['mean_accuracy'],
    color = "blue",
    label="Linear_BoW",
)

plt.plot(
    linear_tfidf_res['C'],
    linear_tfidf_res['mean_accuracy'],
    color = "red",
    label="Linear_TF-IDF"
)

plt.legend(loc="upper right")
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Pelatihan Tahap 2 \nKernel Linear: C vs. Akurasi")
plt.show()
```

***

```python
plt.plot(
    range(len(rbf_bow_res)),
    rbf_bow_res['mean_accuracy'],
    color="blue",
    label="RBF_BoW",
)

plt.plot(
    range(len(rbf_bow_res)),
    rbf_tfidf_res['mean_accuracy'],
    color="red",
    label="RBF_TF-IDF"
)

plt.legend(loc="lower left")
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title("Pelatihan Tahap 2 \nKernel RBF: Iterasi vs. Akurasi")
plt.show()
```

```python
plt.plot(
    range(len(rbf_bow_res)),
    rbf_bow_res['C'],
    color="#111",
    label="RBF_BoW dan RBF_TFIDF"
)

plt.legend(loc="upper left")
plt.xlabel('Iteration')
plt.ylabel('C')
plt.title("Pelatihan Tahap 2 \nKernel RBF: Iterasi vs. C")
plt.show()
```

```python
plt.plot(
    range(len(rbf_bow_res)),
    rbf_bow_res['gamma'],
    color="#111",
    label="RBF_BoW dan RBF_TF-IDF"
)

plt.legend()
plt.xlabel('Iteration')
plt.ylabel('gamma')
plt.title("Pelatihan Tahap 2 \nKernel RBF: Iterasi vs. gamma")
plt.show()
```

***

#### __Poly__

```python
plt.plot(
    range(len(poly_bow_res)),
    poly_bow_res['mean_accuracy'],
    color="blue",
    label="Poly_BoW"
)

plt.plot(
    range(len(poly_bow_res)),
    poly_tfidf_res['mean_accuracy'],
    color="red",
    label="Poly_TF-IDF"
)

plt.legend(loc="lower right")
plt.xlabel('iterasi')
plt.ylabel('akurasi')
plt.title("Pelatihan Tahap 2 \nKernel Polinomial: Iterasi vs. Akurasi")
plt.show()
```

```python
plt.plot(
    range(len(poly_bow_res)),
    poly_bow_res['C'],
    color="#000",
    label="Poly_BoW & Poly_TF-IDF"
)

plt.legend()
plt.xlabel('iterasi')
plt.ylabel('C')
plt.title("Pelatihan Tahap 2 \nKernel Polinomial: Iterasi vs. C")
plt.show()
```

```python
plt.plot(
    range(len(poly_bow_res)),
    poly_bow_res['gamma'],
    color="#111",
    label="Poly_BoW & Poly_TF-IDF"
)

plt.legend()
plt.xlabel('iterasi')
plt.ylabel('gamma')
plt.title("Pelatihan Tahap 2 \nKernel Polinomial: Iterasi vs. gamma")
plt.show()
```

```python
plt.plot(
    range(len(poly_bow_res)),
    poly_bow_res['degree'],
    color="#111",
    label="Poly_BoW & Poly_TF-IDF"
)

plt.legend()
plt.xlabel('iterasi')
plt.ylabel('Degre')
plt.title("Pelatihan Tahap 2 \nKernel Polinomial: Iterasi vs. Degree")
plt.show()
```
